## Criando Novas Variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep=';') # Importando o arquivo csv
dados.head() # Exibindo as primeiras linhas do arquivo
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU'] # Criando nova coluna
dados.head() # Exibindo as primeiras linhas do arquivo
```

```python
dados['Valor M²'] = dados['Valor'] / dados['Area'] # Criando nova coluna
dados.head() # Exibindo as primeiras linhas do arquivo
```

```python
dados['Valor M²'] = dados['Valor M²'].round(2) # Arredondando o valor da coluna
dados.head() # Exibindo as primeiras linhas do arquivo
```

```python
dados['Valor Bruto M²'] = (dados['Valor Bruto'] / dados['Area']).round(2) # Criando nova coluna
dados.head() # Exibindo as primeiras linhas do arquivo
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila'] # Criando uma lista
```

```python
dados['Tipo agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento') # Criando nova coluna
dados.head() # Exibindo as primeiras linhas do arquivo
```

## Excluindo Variáveis

```python
dados_aux = dados[['Tipo agregado', 'Valor M²', 'Valor Bruto', 'Valor Bruto M²']] # Criando uma nova tabela
dados_aux.head() # Exibindo as primeiras linhas do arquivo
```

```python
del dados_aux['Valor Bruto'] # Deletando a coluna
dados_aux.head() # Exibindo as primeiras linhas do arquivo
```

```python
dados_aux.pop('Valor Bruto M²') # Deletando a coluna
dados_aux.head() # Exibindo as primeiras linhas do arquivo
```

```python
dados.drop(['Valor Bruto', 'Valor Bruto M²'], axis=1, inplace=True) # Deletando a coluna
dados.head() # Exibindo as primeiras linhas do arquivo
```

```python
dados.to_csv('dados/aluguel_residencial_aux.csv', sep=';', index=False) # Exportando o arquivo csv
```
